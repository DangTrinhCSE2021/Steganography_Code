```python
import matplotlib.pyplot as plt

def plot_losses(unet,resnet, alexnet):
    plt.figure(figsize=(8, 5))
    plt.plot(range(1, len(unet) + 1), unet, marker='o', linestyle='-',c='red',label="UNet")
    plt.plot(range(1, len(resnet) + 1), resnet, marker='o', linestyle='-',c='blue',label="ResNet")
    plt.plot(range(1, len(alexnet) + 1), alexnet, marker='o', linestyle='-',c='yellow',label="AlexNet")
    
    plt.title('Training Loss Over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.legend()
    plt.show()

# Plot the losses
plot_losses(unet,resnet,alexnet)
```
